# Funkce input
- Aby mohl uživatel zadávat do programu nějaké vstupní informace, tak musí použít funkci input.
- Tato funkce předá tento text do zvolené proměnné.
```

```
- Při spuštění kódu vyskočí okémko, do kterého musíme napsat text. Dokud ho nenapíšeme a neodklikneme enterem, tak nemůžeme pokračovat dál.
```

```
- Do proměnné se nám uložil text, zadaný inputem. Pomocí funkce print se přesvědčíme co je v proměnné uloženo.



````
jmeno = input("Zadej jméno: ")
print(jmeno)
````

# Kalkulacka
- U programování kalkulačky nastává jeden menší problém! Všechno co zadáte do funkce input() se stane textem! Tedy: "5", "6", "7" (Cokoliv je v úvozovkách tak se pro python stává textem)
```
# "5"  + "4" = "54" #Text se spojí dohromady
```
- Jak tedy nenapsat program, který by měl sečíst dvě čísla?

````
print("Dnes bude vyuzivat nasi drobnou kalkulacku!")
cislo1 = input("Zadej cislo1: ")#vzdycky leze str(text)
cislo2 = input("Zadej cislo2: ")
vysledek = cislo1 + cislo2
print("Priklad vysel: ", vysledek)
````

# Python je dynamický jazyk
- Kdykoliv během programu si mohu přepsat to co mám uloženo v proměnné.
- nikdy nepřepisujme jména funkcí, nebo vyhrazené výrazy.
```
print = "text"
```
V tuto chvíli již nemůžeš zavolat funkci print!
```
print("text")
TypeError: 'str' object is not callable
```
- Pokud se dostaneš do této situace tak nezoufej.
- Napiš následující příkaz a všechno bude fungovat zase tak jako má:
```
del print
```








````
print = "text"
````

````
print("Tohle je text")
````

````
del print
````
