# Zkusme zapojit nové znalosti
- Pokud budu chtit zapojit dalsi podminku, tak musim davat elif a ne dalsi if!


```
elif <Pravda>:
<TABULATOR> print("Pro dalsi podminky potrebujeme elif")

```



````
cislo1 = input("Zadej cislo1: ")
cislo1 = int(cislo1) if cislo1.isdecimal() else print("Zkus znovu")
cislo2 = input("Zadej cislo2: ") 
cislo2 = int(cislo2) if cislo2.isdecimal() else print("Zkus znovu")

if cislo1 > cislo2:
  print("cislo1 je > nez cislo2")
elif cislo1 == cislo2:  #kolik chci elif
  print("cislo1 a cislo2 jsou rovny")
elif cislo1 < cislo2: 
  print("cislo2 je > nez cislo1")
else:
  print("Neco jsi spatne zadal")
````

# kod kalkulačky

````
oddelovac = 60 * "="
print(oddelovac)

#Akcni cast
jmeno = input("Zadej jmeno: ")
print(oddelovac)
print("Ahoj " + jmeno  + ", tohle je moje aplikace kalkulacka")
print(oddelovac)
print("""Tady mas na vyber co muzes pocitat:
        1) Scitani  pomoci +
        2) Odcitani pomoci -
        3) Nasobeni pomoci *
        4) Deleni   pomoci / """)


#Akcni cast
print(oddelovac)
cislo1 = float(input("Zadej cislo1: "))
print(oddelovac)
cislo2 = float(input("Zadej cislo 2: "))
print(oddelovac)
operace = input("Zadej operaci z nabidky: ")
print(oddelovac)

#Logicka cast
if operace == "+":
  vysledek = cislo1 + cislo2
elif operace == "-":
  vysledek = cislo1 - cislo2
elif operace == "*":
  vysledek = cislo1 * cislo2
elif operace == "/":
  vysledek = cislo1 / cislo2
else:
  print("Nic se neprovedo")

#Zhodnoceni vysledku
print("Tvůj výsledek je: ", vysledek, "Měj se hezky", jmeno)
print(oddelovac)
````
