```
my_dict = {}
for i in range(10):
  my_dict[f"film{i}"] = {"jmeno": input("Zadej jmeno filmu: "), 
                         "Nejlepší film": input("Kolikaty je to film na CSFD? "), 
                         "Hraji":input("Kdo hraje? Zadej herce a rozdel to carkou").split(",")}
  chces_zadat = input("Chces, zadt dalsi film? Pokud ne, tak zmackni(q)")                       
  if chces_zadat == "q":
    break
  else:
    print(f"Pokracuji a zadavam: {len(my_dict) + 1} film z 10")
```
