```python
import pandas as pd
import sqlite3

data = pd.read_csv("/content/titanic (1).csv")
data.head(2)
```

```python
columns = ", ".join(list(data.columns))
```

```python
conn = sqlite3.connect("Titanic_database.db")
c = conn.cursor()
c.execute(f"CREATE TABLE TITANIC ({columns})")
conn.commit()
```

```python
dir(pd)
```

```python
dir(data)
```

```python
data.to_sql("TITANIK", conn, if_exists='replace', index= False)
```

```python
query = """
        SELECT * 
        FROM TITANIK
        """
pd.read_sql_query(query, con = conn)
```

```python
sql_query = """
            SELECT Sex, Pclass
            FROM TITANIK
            WHERE Survived = 1      
            """
Survived = pd.read_sql_query(sql_query, con = conn)
Survived
```

```python
sql_query = """
            SELECT Sex, Pclass
            FROM TITANIK
            WHERE Survived = 1 AND Pclass = 1     
            """
Survived = pd.read_sql_query(sql_query, con = conn)
Survived
```

```python
sql_query = """
            SELECT Sex, Pclass
            FROM TITANIK
            WHERE Survived = 1 AND Pclass = 1  AND Sex = "female"   
            """
Survived = pd.read_sql_query(sql_query, con = conn)
Survived
```

```python
sql_query = """
            SELECT Name, Sex, Pclass, Age
            FROM TITANIK
            WHERE Survived = 1 AND Pclass = 1  AND Sex = "female"  AND Age < 18  
            """
Survived = pd.read_sql_query(sql_query, con = conn)
Survived
```
